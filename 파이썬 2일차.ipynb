```python
# 데이터 타입

# 숫자
# 문자 : 따음표로 묶기
# 참/거짓(bool / boolean) : True 또는 False

# 연산자
# 대입연산 (=) : 변수, 데이터를 넣는다.

# # 사직연산
  # + : 더하기
  # - : 빼기
  # * : 곱하기
  # / : 나누기
  # % : 나누기 나머지

  # 논리연산
    # == : 같다
    # != : 같지 않다
    # >, < : 초과 또는 미만
    # >=, <= : 이상 또는 이하
    # and : 두 논리가 모두 맞을때만 참
    # or : 두 논리 중 하나만 맞을때 참
a = 5 #대입연산자 ->데이터 넣는다 -> 5데이터를 a 변수에 넣는다.
print(20 + 10)
print(20 - 10)
print(20 * 10)
print(20 / 10)
print(20 % 10)

# 논리연산
print(10 == 10)
print(10 != 10)
print(10 > 10)
print(10 >= 10)
print(10 == 10 and 5 == 5)
# print(10 == 10 and == 5) 논리 연산 X 틀린 문법
print(10 == 10 or 5 != 5)
```

```python
# 조건문

# if 논리연산 :
#   실행코드

# 논리연산이 '참' 인경우만 실행코드 실행. '거짓' 이면 실행을 하지 않는다.
# 들여쓰기 : 조건문 안에 해당 된다
if 10 == 9:
  print(10)

# if 10 == 10:
# print(10)<-들여쓰기를 안해서 실행불가

name = "홍길동"
if name == "홍길동":
  print("맞다")

if name == "김철의": #<- false 라 실행 안함
  print("김철의야")
```

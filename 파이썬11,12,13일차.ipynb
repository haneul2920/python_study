```python
# 인스턴스 속성 은닉

class Person:
  def __init__(self, in_name, in_age, in_weight):
    print('init 실행')
    self.name = in_name
    self.age = in_age
    self.__weight = in_weight

jin = Person('진', 22, 74)
print(jin.name)
# print(jin.weight)

# 인스턴스 속성 접근 : 특정메소드 필요
class Person:
  def __init__(self, in_name, in_age, in_weight):
    print('init 실행')
    self.name = in_name
    self.age = in_age
    self.__weight = in_weight

  def getWeight(self):
    return self.__weight

  @property
  def weight(self):
    return self.__weight


  def setWeight(self, weight):
    self.__weight = weight

  @weight.setter
  def weight(self, weight):
    self.__weight = weight

jin = Person('진', 22, 74)
print(jin.name)
# jin.weight -= 65
print(jin.getWeight()) # 74
jin.setWeight(65)
print(jin.getWeight()) # 65

# 프로퍼티, 세터 어노테이션 사용법
print(jin.weight)
jin.weight = 80
print(jin.weight)
```

```python
# 클래스 속성 은닉
class Person:
  name = '홍길동'
  age = 20
  __weight = 74

a = Person()
# print(a.__weight)

# 은닉된 속성 접근
class Person:
  name = '홍길동'
  age = 20
  __weight = 74

  def getWeight(self):
    return self.__weight

  @property
  def weight(self):
    return self.__weight


  def setWeight(self, weight):
    self.__weight = weight

  @weight.setter
  def weight(self, weight):
    self.__weight = weight

jin = Person()
print(jin.weight) # 74
jin.weight = 66
print(jin.weight) # 66
```

```python
# 클래스 속성은 인스턴스가 공유할 수 있다. 단, 메소드를 활용한다.
class Folder:
  class_folder = []

  def add_file(self, file):
    Folder.class_folder.append(file)

f = Folder()
f.add_file('txt')
f2 = Folder()
f2.add_file('pdf')

print(f.class_folder) # txt, pdf
print(f2.class_folder) # txt, pdf

f.class_folder = 'csv'
f2.class_folder = 'jpg'
print(f.class_folder) # csv
print(f2.class_folder) # jpg
```

```python
# 생성자로 만든 인스턴스 속성 공유X
class Folder:
  def __init__(self):
    self.instance_folder = []

  def add_file(self, file):
    self.instance_folder.append(file)

f = Folder()
f.add_file('txt')
f2 = Folder()
f2.add_file('pdf')

print(f.instance_folder) # txt
print(f2.instance_folder) # pdf

f.class_folder = 'csv'
f2.class_folder = 'jpg'
print(f.instance_folder) # csv
print(f2.instance_folder) # jpg
```

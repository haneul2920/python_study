```python
'''
문제 : 딕셔너리에 각 달의 마지막 날들을 한번에 담기
조건 : 1월 ~ 3월까지
조건 : 2월의 마지막은 28일로 함
조건 : 첫 데이터의 key는 "1월", value는 31
'''
days = {"1월" : 31,
        "2월" : 28,
        "3월" : 31
        }
```

```python
# 문제 : 딕셔너리에 각 달의 마지막 날들을 반복문을 통해 담기
month_last = {}
for i in range(1,13):
  month = str(i) + "월"
  last_day = 31
  if i == 2:
    last_day = 28
  if i == 4 or i == 6 or i == 9 or i == 11:
    last_day = 30

  month_last[month] = last_day

print(month_last)
```

```python
# 문제 : 딕셔너리에 각 달의 마지막 날들을 한번에 담고, 수작업으로 순회출력
month_last = {"1월" : 31,
              "2월" : 28,
              "3월" : 31}

for i in month_last:
  print(i,month_last[i])
```

```python
# 문제 : 딕셔너리에 각 달의 마지막 날들을 한번에 담고, keys()로 key 순회출력
month_last = {"1월" : 31,
              "2월" : 28,
              "3월" : 31}

for i in month_last.keys():
  print(i)
```

```python
# 문제 : 딕셔너리에 각 달의 마지막 날들을 한번에 담고, keys()로 key, value 순회출력
month_last = {"1월" : 31,
              "2월" : 28,
              "3월" : 31}

for i in month_last.keys():
   print(i, month_last[i])
```

```python
# 문제 : 딕셔너리에 각 달의 마지막 날들을 한번에 담고, values()로 value 순회출력

month_last = {"1월" : 31,
              "2월" : 28,
              "3월" : 31}

for i in month_last.values():
  print(i)
```

```python
# 문제 : 딕셔너리에 각 달의 마지막 날들을 한번에 담고, items()로 key, value 순회출력
month_last = {"1월" : 31,
              "2월" : 28,
              "3월" : 31}

for i,j in month_last.items():
  print(i,j)
```

```python
print("== 예제 1 ==")
s = "안녕하세요."

print("s의 길이 출력")
# 구현
print(len(s))

print("s를 이루는 문자 중에서 앞에서 문자 2개 출력")
# 구현
print(s[0:2])

print("s를 이루는 문자 중에서 앞에서 문자 하나 건너 뛰고 1개 출력")
# 구현
print(s[1:2])

print("s를 이루는 문자 중에서 앞에서 문자 4개 출력, 첫 인덱스 생략")
# 구현
print(s[:4])

print("s를 이루는 문자 중에서 앞에서 문자 3개 건너뛰고, 나머지 전부 출력, len")
# 구현
print(s[len(s)-3::])

print("s를 이루는 문자 중에서 앞에서 문자 3개 건너뛰고, 나머지 전부 출력, 뒤 인덱스 생략, 방법 2")
# 구현
print(s[3::])

print("== 예제 2 ==")
s = "안녕하세요"

print("s를 이루는 문자를 하나씩 출력, while")
# 구현
i = 0
while i < len(s):
  print(s[i])
  i += 1

print("s를 이루는 문자를 역순으로 하나씩 출력, while")
# 구현
i = 1
while i < len(s)+1:
  print(s[-i])
  i += 1

print("s를 이루는 문자를 하나씩 출력, for")
# 구현
for i in range (0,5):
  print(s[i])

for j in s:
  print(j)

print("== 예제 3 ==")
s = "a안b녕c하d세e요f"

print("s를 이루는 문자 중에서 한글만 출력")
# 구현
print(s[1::2])
print(s[1]+s[3]+s[5]+s[7]+s[9])
print("s를 이루는 문자 중에서 한글만 출력, for, if or, continue")
# 구현
answer = ""
for i in range(1,10,2):
  answer += s[i]
print(answer)
answer2 = ""
for x in s:
  if x == "안" or x == "녕" or x == "하" or x == "세" or x == "요":
    answer2 += x
print(answer2)

answer3 = ""
for x in s:
  if x == "a" or x == "b" or x == "c" or x == "d" or x == "e" or x == "f":
    continue
  answer3 += x
print(answer3)


print("s를 이루는 문자 중에서 한글만 출력, for, if and, 방법 2")
# 구현
answer4 = ""
for i in s:
  if i != 'a' and i != 'b' and i != 'c' and i != 'd' and i != 'e' and i != 'f':
    answer4 += i
print(answer4)

print("s를 이루는 문자 중에서 한글만 출력, for, not in, 방법 3")
# 구현
answer5 = ""
for i in s:
  if i not in ['a','b','c','d','e','f']:
    answer5 += i
print(answer5)
```

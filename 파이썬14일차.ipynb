```python
# 메소드 / 메서드 method
# 1. 인스턴스 메소드
class Sport:
  def run(self, km, dis):
    return 'run', km, dis

a = Sport()
print(a.run(10,20)) # 인스턴스 메소드

# 2. 스테틱 메소드(정적 메소드)
class Calc:
  @staticmethod
  def add(a,b):
    return a + b

class Calc2:
  @staticmethod
  def add(a,b):
    return a - b

print(Calc.add(10,20))
print(Calc2.add(10,20))

# 3. 클래스 메소드
class Person:
  count = 0

  @classmethod
  def print_count(cls):
    print('클래스 메소드 실행 : ',cls.count)
    cls.count += 1

Person.print_count()
Person.print_count()
Person.print_count()
```

```python
class Person:
  name = 'john'

  @classmethod
  def class_print(cls):
    print('클래스 메소드 실행 : ',cls.name)

  @staticmethod
  def static_print():
    print('클래스 메소드 실행 : 클래스 속성에 접근 불가')

print(Person.name)
Person.class_print()
Person.static_print()
```

```python
# 인스턴스 속성 물려받기2 : super().__init__()
class Person:
  def __init__(self):
    self.hello = 'hi'

class Student(Person):
  def __init__(self):
    super().__init__()

a = Student()
print(a.hello)
```

```python
# 인스턴스 다중상속(파이썬만의 문법)
class Person:
  def __init__(self):
    self.hello = 'hi'
    super().__init__()

class Person2:
  def __init__(self):
    self.hello2 = 'hi2'

class Student(Person,Person2):
  def __init__(self):
    self.student = 'student attr'
    # Person.__init__(self)
    # Person2.__init__(self)
    super().__init__()

a = Student()
print(Student.__mro__) # MRO(Method Resolution Order)
print(a.student)
print(a.hello)
print(a.hello2)
```

```python
# 메소드 '자녀(부모)' 방식 상속
class Person:
  def hello(self):
    print('instance hello')

  @staticmethod
  def static_hello():
    print('static hello')

  @classmethod
  def class_hello(cls):
    print('class {} hello'.format(cls.__name__))
    print('attr {}'.format(cls.study))

class Student(Person):
  study = 'hard'

class Worker(Person):
  pass

a = Student()
a.hello()
a.static_hello()
a.class_hello()
b = Worker()
b.hello()
b.static_hello()
b.class_hello()

print()
# Student.hello() # 오류. 인스턴스 만들어서 써야한다.
Student.static_hello()
Student.class_hello()
Worker.static_hello()
Worker.class_hello()
```

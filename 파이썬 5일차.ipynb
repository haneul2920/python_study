```python
# 함수 : 매개변수, 인자

# 1. in, out X
def print_1_to_10():
  i=1
  while i <= 10:
    print(i)
    i += 1

# 2. in X, out O
def get123():
  return 123

# 3. in O, out X
def print_1_to_m(m):
  i = 1
  while i <= m:
    print(i)
    i += 1

get123()
print_1_to_10()
print_1_to_m(5)

# 4. in O(복수), out X
def print_n_to_m(n,m):
  i = n
  while i <= m:
    print(i)
    i+=1

print_n_to_m(2,5)
print_n_to_m(10,15)
print_n_to_m(22,25)
print_n_to_m(32,35)

# 5. in O, out O
def sum_n_to_m(n,m):
  sum = 0
  i = n
  while i <= m:
    sum += i
    i +=1
  return sum

print(sum_n_to_m(1,5))
print(sum_n_to_m(10,15))
```

```python
# 문제 : 매개변수의 개수를 맞춰주세요.
# 조건 : 함수호출 시 사용한 인자에 따라 함수를 적절하게 잘 만들어 주세요.
# 조건 : 오류가 나지 않으면 성공입니다.

def a():
  print()

def b(a,b,c):
  print(a,b,c)

def c(z,a,b,c):
  print("{},{},{},{}".format(z,a,b,c))




a()
b(1, 2, 3)
c("안녕", 1 == 1, 550, 600)
```

```python
# 문제 : 입력받은 정수의 모든 약수를 출력하는 함수를 구현해주세요.

def print_divisors(num):
  i = 1
   # 구현
  while i <= num:
    if num % i == 0:
      print(i)
    i += 1





print_divisors(1000)
print_divisors(50)
```

```python
# 전역변수, 지역변수
a = 1

def b():
  a = 2
  print("함수 내 a : ",a) # 2

print("함수 실행 전 a : ",a) # 1
b()
print("함수 실행 후 a : ",a) # 1
a=3
print("함수 수정 후 a : ",a) # 3
```

```python
# 지역변수를 전역변수로 사용하는방법
a = 1#전역변수

def b():
  global a #전역변수로 인식
  a = 2
  print("함수 내 a : ",a) # 2

print("함수 실행 전 a : ",a) # 1
b()
print("함수 실행 후 a : ",a) # 1
a=3
print("함수 수정 후 a : ",a) # 3
```

```python
import pygame

# 1. 초기화 -> 변수의 선언
pygame.init()

# 2. 게임 화면 설정 -> 고정값
size = [600,800]
screen = pygame.display.set_mode(size)
title = 'pyamge example'
pygame.display.set_caption(title)

# 3. 게임 내에서의 설정 -> 변수

clock = pygame.time.Clock()
color1 = (83,104,154)

class Img:
    def __init__(self):
        self.x = 0
        self.y = 0
        self.speed = 0


    def img_load(self, address):
        self.img = pygame.image.load(address)

    def change_size(self, width, height):
        self.img = pygame.transform.scale(self.img,(width, height))
        self.width = width
        self.height = height

    def show_img(self):
        screen.blit(self.img,(self.x, self.y))

    # v.2
    # def set_speed(self, speed)
    #     self.speed = speed

hero = Img()
hero.img_load('C:/Users/admin/Desktop/pygame/spaceship.png')
hero.change_size(100,100)
hero.x = (size[0]/2) - (hero.width/2)
hero.y = size[1] - hero.height - 30
hero.speed = 5
# hero.set_speed(5)

enemy = Img()
enemy.img_load('C:/Users/admin/Desktop/pygame/images.png')
enemy.change_size(80,80)
enemy.x = 300
enemy.y = 10



# enemy = pygame.image.load('C:/Users/admin/Desktop/pygame/images.png')
# enemy = pygame.transform.scale(enemy,(80,80))
# enemy_width, enemy_height = enemy.get_size()

# hero = pygame.image.load('C:/Users/admin/Desktop/pygame/spaceship.png')
# hero = pygame.transform.scale(hero,(100,100))
# hero_width, hero_height = hero.get_size()


right_move = False
left_move = False
up_move = False
down_move = False


# 4. 메인 이벤트
system = 0
while system == 0:

#  -1. FPS(Frame per sec) 설정
    clock.tick(60)

#  -2. 입력(키보드,마우스)의 감지 -> 활용
    for event in pygame.event.get():
        # print("평상시 : ",event)
        if event.type == pygame.QUIT:
            system = 1

        if event.type == pygame.KEYDOWN:
            if event.key == pygame.K_LEFT:
                left_move = True
            if event.key == pygame.K_RIGHT:
                right_move = True
            if event.key == pygame.K_UP:
                up_move = True
            if event.key == pygame.K_DOWN:
                down_move = True

        if event.type == pygame.KEYUP:
            if event.key == pygame.K_LEFT:
                left_move = False
            if event.key == pygame.K_RIGHT:
                right_move = False
            if event.key == pygame.K_UP:
                up_move = False
            if event.key == pygame.K_DOWN:
                down_move = False

    if right_move == True:
        hero.x += hero.speed
    if left_move == True:
        hero.x -= hero.speed
    if up_move == True:
        hero.y -= hero.speed
    if down_move == True:
        hero.y += hero.speed

#  -3. 입력에 따른 변화, 시간에 따른 변화
    if hero.x < 0:
        hero.x = 0
    if hero.x > size[0] - hero.width:
        hero.x = size[0] - hero.width
    if hero.y < 0:
        hero.y = 0
    if hero.y > size[1] - hero.height:
        hero.y = size[1] - hero.height


#  -4. 전사작업(그리기)
    screen.fill(color1)
    hero.show_img()
    enemy.show_img()
    # screen.blit(hero, (hero_x, hero_y))
    # screen.blit(enemy,(10,10))
#  -5. 업데이트
    pygame.display.flip()
# 5. 종료(x버튼, game over)
pygame.quit()
```

```python
# 문제 : 리스트에 각 달의 끝 날짜들을 담고, '1월은 31일까지'와 같은 양식으로 출력
a = [31,29,31,30,31,30,31,31,30,31,30,31]

count = 0
for i in range(1, len(a)+1, 1):
  print("{}월은 {}일까지".format(i,a[count]))
  count += 1
```

```python
# 문제 : 리스트에 각 달의 끝 날짜들을 담고, '1월은 31일까지'와 같은 양식으로 출력, len 사용
a = [31,29,31,30,31,30,31,31,30,31,30,31]

count = 0
for i in range(1, len(a)+1, 1):
  print("{}월은 {}일까지".format(i,a[count]))
  count += 1
```

```python
# 문제 : 리스트에 각 달의 끝 날짜들을 담고, '1월은 31일까지'와 같은 양식으로 출력, enumerate 사용

# enumerate : <index, 요소> 모두 출력

a = [31,29,31,30,31,30,31,31,30,31,30,31]

for i,a in enumerate(a):
  print("{}월은 {}일까지".format(i+1,a))
```

```python
# 별 피라미드를 출력해주세요.
# i 변수에 넣는 숫자는 피라미드의 층수를 의미합니다.

i = 5

for a in range(0, i,1):
  for b in range(0, i-a, 1):
    print("/",end = "")
  for c in range(-1, a*2, 1):
    print("*",end = "")
  print()




#아래와 같이 출력되도록 코드를 작성해보세요.
# '''
#     *
#    ***
#   *****
#  *******
# *********
# '''
```

```python
# 숫자 피라미드 버전1을 출력해주세요.
# i 변수에 넣는 숫자는 피라미드의 층수를 의미합니다.

# 1. 틀잡기
# i = 5
# for level in range(1, i + 1):
#   for 공백
#   for 왼편 숫자
#   for 오른편 숫자
#   print()

# 2.
# v1 층수 반복코드를 상행
# i = 9 # 층수 조정 변수
# for level in range(1, i + 1): # 피라미드 층수
#   for b in range(i - level, -1, -1): # 공백
#     print(" ", end="")
#   for left_num in range(i - level + 1, i, 1): # 왼편 숫자
#     print(left_num, end="")
#   for right_num in range(i, i - level, -1): # 오른편 숫자
#     print(right_num, end="")
#   print()

# v2 층수 코드를 하행
# 1.틀잡기
i = 5
for level in range(i,0,-1):
  for b in range(level - 1, 0, -1):
    print(" ",end = "")
  for left_num in range(level, i + 1, 1):
    print(left_num,end = "")
  for right_num in range(i - 1, level - 1,-1):
    print(right_num,end = "")
  print()

#아래와 백이 출력되도록 코드를 작성해보세요.
# '''
#     5
#    454
#   34543
#  2345432
# 123454321
# '''
```

```python
# 숫자 피라미드 버전2를 출력해주세요.
# i 변수에 넣는 숫자는 피라미드의 층수를 의미합니다.

i = 5

for level in range(1, i + 1): # 피라미드 층수
  for b in range(i - level, -1, -1): # 공백
    print(" ", end="")
  for left_num in range(level, 1, -1): # 왼편 숫자
    print(left_num, end="")
  for right_num in range(1,level+1, 1): # 오른편 숫자
    print(right_num, end="")
  print()

#아래와 같이 출력되도록 코드를 작성해보세요.
# '''
#     1
#    212
#   32123
#  4321234
# 543212345
# '''
```

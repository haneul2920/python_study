```python
import openpyxl as op
wb = op.load_workbook(r'C:\Users\admin\Desktop\엑셀자동화_yhw\load3.xlsx')
ws = wb.active

# 시트.rows / 시트.columns
for row in ws.rows:
    print(row)
for col in ws.columns:
    print(col)

for row in ws.rows:
    for cell in row:
        print(cell.value)

# 시트.max_row / 시트.max_column
for row_index in range(1, ws.max_row + 1):
    print(ws.cell(row = row_index, column = 1).value)


wb.save(r'C:\Users\admin\Desktop\엑셀자동화_yhw\load3.xlsx')
```

```python
import openpyxl as op
wb = op.load_workbook(r'C:\Users\admin\Desktop\엑셀자동화_yhw\load3.xlsx')
ws = wb.active

total_row = ws.max_row

# H열에 sum함수 넣기
for row in range(1, total_row + 1):
    ws['H' + str(row)] = f'=SUM(A{row}:F{row})'

wb.save(r'C:\Users\admin\Desktop\엑셀자동화_yhw\load3.xlsx')
```

```python
import openpyxl as op
wb = op.load_workbook(r'C:\Users\admin\Desktop\엑셀자동화_yhw\load3.xlsx')
ws = wb.active

total_row = ws.max_row
for row in range(1, total_row + 1):
    ws['I' + str(row)] = '=A' + str(row) + '*' + 'B' + str(row)

wb.save(r'C:\Users\admin\Desktop\엑셀자동화_yhw\load3.xlsx')
```

```python
import openpyxl as op
wb = op.load_workbook(r'C:\Users\admin\Desktop\엑셀자동화_yhw\load3.xlsx', data_only = True)
ws = wb.active

# 파이썬에서 함수 처리
total_row = ws.max_row
for row in range(1, total_row+1):
    data = ws['A' + str(row)].value
    print(data)
    if data > 1000 :
        ws['H' + str(row)] = '합격'
    else:
        ws['H' + str(row)] = '불합격'


wb.save(r'C:\Users\admin\Desktop\엑셀자동화_yhw\load3.xlsx')
```

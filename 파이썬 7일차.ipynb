```python
# list(리스트)

# 변수는 하나의 데이터만 담을 수 있다.
# 여러개의 데이터를 한번에 이동시키면 편할 텐데

# 시나리오 : 인명부(이름, 나이, 번호, 이메일)

name = ["홍길동","홍길순","박철민"]
ages = [10,20,30]

ages2 = ages
ages3 = ages

print("ages : ", id(ages))
print("ages2 : ", id(ages2))
print("ages3 : ", id(ages3))

print(id([10,20,30]))

print(ages[0])
print(ages2[0])
print(ages3[0])


# 인덱스 : 데이터의 위치, 자동부여

print(name[0])
print(ages[0])
```

```python
# 리스트 활용

# 데이터 추가 / 읽기 / 수정 / 삭제 (feat.CRUD : create, read, update, delete)

# 객체 생성
obj = []
print("객체 생성 : ", obj)
# 추가 : .append()
obj.append(123)
print("데이터 추가 후: ",obj)
obj.append(888)
obj.append(2)
obj.append(56)
print("데이터 추가 후: ",obj)


# 수정 : 인덱스 찾아서 변수 수정하듯
# 인덱스는 리스트 데이터 방의 번호. 0부터 자동부여
obj[0] # 리스트의 0번째 데이터를 가져온다
print("obj 0번 인덱스의 데이터 : ",obj[0])

obj[0] = 99 # 0번째 데이터 수정
print("obj 0번 인덱스의 데이터 : ",obj[0])
print(obj)


# 삭제 : .remove(), del
# 삭제를 하면 남은 데이터들은 0번 인덱스부터 재정렬

obj.remove(99) # 어떤 데이터를 삭제할 지 결정
print("obj 99 삭제 후 : ", obj)

del obj[3] # 인덱스로 찾은 데이터 삭제
print("obj 3번 인덱스 삭제 후 : ", obj)
```

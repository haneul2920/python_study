```python
# 문제 - 1부터 100까지의 수 중에서 짝수의 합을 출력해주세요.(개념 : while, if, %)
sum = 0
i = 1

while i <= 100:
  if i % 2 == 0:
    sum += i
  i += 1
print(sum)
```

```python
# 문제 - 1부터 100까지의 수 중에서 짝수이고 3의 배수가 아닌 수의 합을 출력해주세요.(개념 : while, if, %)
sum = 0
i = 1
while i <= 100:
  if i % 2 == 0 and i % 3 != 0:
    sum = sum + i
  i += 1
print(sum)
```

```python
# 문제 - 구구구단을 출력해주세요.
# 조건 - 3단까지만 출력해주세요.
# 출력양식
'''
== 1단 ==
1 * 1 * 1 = 1
1 * 1 * 2 = 2
1 * 1 * 3 = 3
1 * 2 * 1 = 2
1 * 2 * 2 = 4
1 * 2 * 3 = 6
1 * 3 * 1 = 3
1 * 3 * 2 = 6
1 * 3 * 3 = 9
== 2단 ==
2 * 1 * 1 = 2
2 * 1 * 2 = 4
2 * 1 * 3 = 6
2 * 2 * 1 = 4
2 * 2 * 2 = 8
2 * 2 * 3 = 12
2 * 3 * 1 = 6
2 * 3 * 2 = 12
2 * 3 * 3 = 18
== 3단 ==
3 * 1 * 1 = 3
3 * 1 * 2 = 6
3 * 1 * 3 = 9
3 * 2 * 1 = 6
3 * 2 * 2 = 12
3 * 2 * 3 = 18
3 * 3 * 1 = 9
3 * 3 * 2 = 18
3 * 3 * 3 = 27
'''
k = 1
while k <= 3:
  j = 1
  print(f"=={k}단==")
  while j <= 3:

    i = 1
    while i <= 3:
      print("{} X {} X {} = {}".format(k,j,i,k*j*i))
      i += 1

    j += 1

  k += 1


```
